### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
```

```python
df['Species'] = lr.fit_transform(df['Species'])
df.head()
```

### Model Training

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['Species'])
y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print("Logistic Regression Accuracy:",model.score(x_test,y_test)*100)
```

```python
lr=model.score(x_test,y_test)*100
lr
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print("KNeighborsClassifier Accuracy:",model.score(x_test,y_test)*100)
```

```python
KNN = model.score(x_test,y_test)*100
KNN
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print("DecisionTreeClassifier Accuracy:",model.score(x_test,y_test)*100)
```

```python
DTC = model.score(x_test,y_test)*100
DTC
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
# Scaling the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

models = {
    "Logistic Regression": LogisticRegression(solver='lbfgs', max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

# Create confusion matrices
confusion_matrices = {}

for model_name, model in models.items():
    model.fit(x_train_scaled, y_train)
    y_pred = model.predict(x_test_scaled)
    confusion_matrices[model_name] = confusion_matrix(y_test, y_pred)

# Display confusion matrices using a heatmap
plt.figure(figsize=(12, 8))
for idx, (model_name, matrix) in enumerate(confusion_matrices.items(), 1):
    plt.subplot(2, 2, idx)
    sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```
